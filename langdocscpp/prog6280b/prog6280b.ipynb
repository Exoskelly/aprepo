```python
%pip install pandas seaborn matplotlib -q
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../Langdat/starbucks.csv')
df.columns = df.columns.str.strip() 
df
```

```python
df.info()
```

```python
df.describe()
```

Task 1

```python
plt.figure(figsize=(10,6))
df.groupby('Beverage_category')['Calories'].mean().plot(kind='barh', color='purple')
plt.show()
```

Task 2

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Total Fat (g)', y='Saturated Fat (g)', data=df) 
plt.title('Total Fat vs. Saturated Fat')
plt.xlabel('Total Fat')
plt.ylabel('Saturated Fat')
plt.show()
```

Task 3

```python
plt.figure(figsize=(10,6))
sns.histplot(y='Caffeine (mg)', data=df.head(2500))
plt.title('Caffeine per (mg)')
plt.ylabel('Caffeine')
plt.show()
```

Task 4

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Beverage_category', y=' Protein (g)', data=df)  # sns.violinplot(...)
plt.title('Protien by Beverage Catagory')
plt.xlabel('Beverage Type')
plt.ylabel('Protein')
plt.show()
```
